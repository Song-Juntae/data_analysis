```python
df["상호명_대문자"] = df["상호명"].str.upper()
df.loc[df["상호명_대문자"].str.contains("버거킹|BKR"), "상호명_대문자"].unique()
```

```python
df[df["상호명_대문자"].str.contains(
    "버거킹|BKR|맥도날드|멕도날드|맥도널드|McDonald|롯데리아|KFC|케이에프씨") 
          & (df["상권업종대분류명"] != "음식")]
```

```python
df_b = df[df["상호명_대문자"].str.contains(
    "버거킹|BKR|맥도날드|멕도날드|맥도널드|McDonald|롯데리아|KFC|케이에프씨") 
          & (df["상권업종대분류명"].isin(["음식", "소매", "생활서비스"]))
         ]
df_b.shape
```

```python
df_b
```

```python
bmkl = {"버거킹" : "버거킹|BKR",
        "맥도날드" : "맥도날드|멕도날드|맥도널드|McDonald",
        "롯데리아" : "롯데리아",
        "KFC" : "KFC|케이에프씨"}
for brand, val in bmkl.items() :
    df_b.loc[df_b["상호명_대문자"].str.contains(val), "브랜드"] = brand
```

```python
df_b
```

```python
df_b["브랜드"].value_counts()
```

```python
df_skorea = pd.crosstab(df_b["시도명"], df_b["브랜드"])
df_skorea["합계"] = df_skorea.sum(axis=1)
df_skorea
```

```python
df_skorea = df_b.pivot_table(index="시도명", columns="브랜드",
                 values="상호명", aggfunc="count",
                 fill_value=0, margins=True, margins_name="합계")
df_skorea
```

```python
df_skorea = df_b.groupby(
    ["시도명", "브랜드"]
)["상호명"].count().unstack().fillna(0).astype(int)
df_skorea["합계"] = df_skorea.sum(axis=1)
df_skorea
```

```python
sns.heatmap(df_skorea.iloc[:, :4], annot=True, cmap="Blues", fmt=".0f")
```
